```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = (r'C:\Users\A454\Desktop\clone\UBHM-VIRT-DATA-PT-05-2022-U-LOL\UBHM-VIRT-DATA-PT-05-2022-U-LOL\05-Matplotlib\Home Work\Instructions\Pymaceuticals\data\Mouse_metadata.csv')
study_results_path = (r'C:\Users\A454\Desktop\clone\UBHM-VIRT-DATA-PT-05-2022-U-LOL\UBHM-VIRT-DATA-PT-05-2022-U-LOL\05-Matplotlib\Home Work\Instructions\Pymaceuticals\data\Study_results.csv')

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path) #249 rows, 5 columns
study_results = pd.read_csv(study_results_path) # 1893 rows, 4 columns

# Combine the data into a single dataset
combined_df = pd.merge(mouse_metadata, study_results)
# Display the data table for preview
#combined_df.head()
combined_df #1893 rows, 8 columns
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

#group the clean & combined data frame by Drug Regimen
by_regimen = clean_combined.groupby("Drug Regimen")

#calculate the mean tumor volume
mean_tum_vol = by_regimen["Tumor Volume (mm3)"].mean()

#begin creating the DataFrame by turning the previous result into a DataFrame
summary_stats = pd.DataFrame(mean_tum_vol)

#add the rest of the columns to the dataframe by calculating each statistic
summary_stats["Median Tumor Vol (mm3)"] = by_regimen["Tumor Volume (mm3)"].median()
summary_stats["Variance Tumor Vol (mm3)"] = by_regimen["Tumor Volume (mm3)"].var()
summary_stats["STDEV Tumor Vol (mm3)"] = by_regimen["Tumor Volume (mm3)"].std()
summary_stats["SEM Tumor Vol (mm3)"] = by_regimen["Tumor Volume (mm3)"].sem()

#view the resulting dataframe
summary_stats.head()


# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.

```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line
clean_combined.groupby("Drug Regimen")["Tumor Volume (mm3)"].aggregate(['mean', np.median, np.var, np.std, st.sem])
```

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
greatest_timepoint_bymouse = combined_df.groupby("Mouse ID")["Timepoint"].aggregate(['max'])
g = greatest_timepoint_bymouse.rename(columns={"max": "Timepoint"})
gtbm = g.reset_index()
gtbm
# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
tumor_last_timepoint = pd.merge(combined_df, gtbm)
tumor_last_timepoint.head()
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumvol = []

    # Locate the rows which contain mice on each drug and get the tumor volumes
capomulin = tumor_last_timepoint.loc[tumor_last_timepoint["Drug Regimen"] == "Capomulin"]
ramicane = tumor_last_timepoint.loc[tumor_last_timepoint["Drug Regimen"] == "Ramicane"]
infubinol = tumor_last_timepoint.loc[tumor_last_timepoint["Drug Regimen"] == "Infubinol"]
ceftamin = tumor_last_timepoint.loc[tumor_last_timepoint["Drug Regimen"] == "Ceftamin"]

#Capomulin
sorted_capomulin = capomulin.sort_values('Tumor Volume (mm3)')
sorted_capomulin.head()
capomulin_quartiles = sorted_capomulin['Tumor Volume (mm3)'].quantile([.25,.5,.75])

capomulin_lowerq = capomulin_quartiles[0.25]
capomulin_upperq = capomulin_quartiles[0.75]
capomulin_iqr = capomulin_upperq-capomulin_lowerq

print(f"Capomulin: Lower quartile is: {capomulin_lowerq}")
print(f"Capomulin: Upper quartile: {capomulin_upperq}")
print(f"Capomulin: Interquartile range: {capomulin_iqr}")
      
capomulin_lower_bound = capomulin_lowerq - (1.5*capomulin_iqr)
capomulin_upper_bound = capomulin_upperq + (1.5*capomulin_iqr)
print(f"Capomulin: Values below {capomulin_lower_bound} could potentially be outliers.")
print(f"Capomulin: Values above {capomulin_upper_bound} could potentially be outliers.")


cap_list = capomulin["Tumor Volume (mm3)"].tolist()


fig1, ax1 = plt.subplots()
ax1.set_title('Final Tumor Volume for Capomulin')
ax1.set_ylabel('Tumor Vol (mm3)')
ax1.boxplot(cap_list)
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

s185.iloc[:,6], s185.iloc[:,4]

weight = s185.iloc[:,4]
tvolume = s185.iloc[:,6]
correlation = st.pearsonr(weight,tvolume)
print(f"The correlation between both factors is {round(correlation[0],2)}")

#we get 'nan' because the weight is constant throughout the experiment (17g)


```
